```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# Assuming df is your DataFrame containing inflation data

# Convert non-numeric values to NaN
df.iloc[:, 1] = pd.to_numeric(df.iloc[:, 1], errors='coerce')
df.iloc[:, 2] = pd.to_numeric(df.iloc[:, 2], errors='coerce')

# Drop rows with NaN values
df.dropna(inplace=True)

# EDA and Visualization
plt.figure(figsize=(12, 6))
plt.plot(df.index, df.iloc[:, 2], label='US Inflation', color='blue')
plt.plot(df.index, df.iloc[:, 1], label='India Inflation', color='red')
plt.title('Inflation Over Time')
plt.xlabel('Date')
plt.ylabel('Inflation')
plt.legend()
plt.show()

# Time series decomposition
us_decomposition = seasonal_decompose(df.iloc[:, 2], model='additive')
india_decomposition = seasonal_decompose(df.iloc[:, 1], model='additive')

# Plot decomposed components
us_decomposition.plot()
plt.title('US Inflation Decomposition')
plt.show()

india_decomposition.plot()
plt.title('India Inflation Decomposition')
plt.show()

# Stationarity test (ADF test)
def adf_test(data):
    result = adfuller(data)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {}'.format(key, value))

print('ADF Test Results for US Inflation:')
adf_test(df.iloc[:, 2])

print('\nADF Test Results for India Inflation:')
adf_test(df.iloc[:, 1])

```
